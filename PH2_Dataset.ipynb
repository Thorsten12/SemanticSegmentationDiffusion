## Quick Visualization

```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

base_path = path

for folder in os.listdir(base_path)[:3]:
    folder_path = os.path.join(base_path, folder)
    
    if os.path.isdir(folder_path):
        bmp_files = []
        
        for root, _, files in os.walk(folder_path):
            for file in files:
                if file.lower().endswith(".bmp"):
                    bmp_files.append(os.path.join(root, file))
        
        if len(bmp_files) >= 2:
            # erste 2 Bilder nehmen
            img1 = np.array(Image.open(bmp_files[0]).convert("RGB"))
            img2 = np.array(Image.open(bmp_files[1]).convert("L"))  # Maske
            
            # Maske: schwarz → transparent
            alpha = (img2 > 0).astype(float) * 0.4  # 0.4 = Transparenz
            
            plt.figure(figsize=(5, 5))
            plt.imshow(img1)
            plt.imshow(img2, cmap="Blues", alpha=alpha)
            plt.title(folder)
            plt.axis("off")
            plt.show()
```
